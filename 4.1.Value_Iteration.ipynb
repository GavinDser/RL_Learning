```python
from GridWorld_V1 import GridWorld_v1
import random
import numpy as np
```

```python
gamma = 0.9
rows = 5
columns = 5
theta = 0.001
fixed = [".....",".xx..","..x..",".x!x.",".x..."]
grid = GridWorld_v1(fixed=fixed)
grid.show_map()

np.set_printoptions(precision=4)

value = np.zeros(rows*columns)
qtable = np.zeros((rows*columns,5))
policy = np.argmax(qtable,axis=1)
grid.show_policy(policy)
```

```python
pre_value = value + 10

grid.show_map()
grid.show_policy(policy)

while (np.sum(pre_value - value)**2) > theta:
    print("Euclidean Distance:",np.sum((pre_value-value)**2))
    pre_value = value.copy()
    
    for i in range(rows * columns):
        currentX = i // columns
        currentY = i % columns
        
        #5 actions
        for j in range(5):
            #because state is set up using the formula 
            score, nextState = grid.step(i,j)
            qtable[i][j] = score + gamma * value[nextState]
            
    policy = np.argmax(qtable,axis=1)
    
    value = np.max(qtable,axis=1)
    
    grid.show_policy(policy)
    
print(policy)
print(value)
```
